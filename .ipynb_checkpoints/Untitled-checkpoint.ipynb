```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
aacn_scores = []
continuous_scores = []

for i in range(8):
    data = np.load(f"./log/data/aacn/trial_{i}.npy")
    aacn_scores.append(data)
    data = np.load(f"./log/data/continuous/trial_{i}.npy")
    continuous_scores.append(data)
```

```python
for i in range(8):
    num = 1
    fig, ax = plt.subplots(1)
    
    X1 = np.array([sum(aacn_scores[i][j:j+num])/num for j in range(1400//num)])
    X2 = np.array([sum(continuous_scores[i][j:j+num])/num for j in range(1400//num)])
    

    sigma1 = np.array([aacn_scores[i][j:j+num] for j in range(1400//num)]).std(axis=1)
    sigma2 = np.array([continuous_scores[i][j:j+num] for j in range(1400//num)]).std(axis=1)
    
    t1 = np.arange(0, len(X1))*num
    t2 = np.arange(0, len(X2))*num
    
    ax.plot(t1, X1, label="aacn", color="blue")
    ax.plot(t2, X2, label="base line", color="orange")
    ax.fill_between(t1, X1+sigma1, X1-sigma1, facecolor='blue', alpha=0.3)
    ax.fill_between(t2, X2+sigma2, X2-sigma2, facecolor='orange', alpha=0.3)
    
    ax.set_xlabel('num episode')
    ax.set_ylabel('score')
    ax.legend()
    
    # plt.savefig(f"./log/figure/scores/result_{i}.png")
    plt.show()
    plt.close()
```

```python
for i in range(8):
    num = 5
    fig, ax = plt.subplots(1)
    
    X1 = np.array([sum(aacn_scores[i][j:j+num])/num for j in range(1400//num)])
    X2 = np.array([sum(continuous_scores[i][j:j+num])/20 for j in range(1400//num)])
    

    # Nsteps length arrays empirical means and standard deviations of both
    # populations over time
    sigma1 = X1.std()
    sigma2 = X2.std()
    
    t1 = np.arange(0, len(X1))
    t2 = np.arange(0, len(X2))
    
    ax.plot(aacn_scores[i], label="aacn", color="blue")
    # ax.plot(continuous_scores[i], label="basic", color="orange")
    
    ax.set_xlabel('num episode')
    ax.set_ylabel('score')
    ax.legend()
    
    plt.show()
    
    plt.close()
```
